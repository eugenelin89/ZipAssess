```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df =pd.read_csv('christopher.csv', header=None)[[3,4,5,7]]
```

```python
df[7] = df[7].astype(int)
```

```python
X = df[[3,4,5]].to_numpy()
y = df[7].to_numpy()
```

```python
logistic_model = LogisticRegression(random_state=0).fit(X, y)
```

```python
table = []
for i in range(1,13):
    for j in range(1, 13):
        table.append([i,j,i*j])
```

```python
result_prediction = logistic_model.predict_proba(np.array(table))
```

```python
result_prediction
```
